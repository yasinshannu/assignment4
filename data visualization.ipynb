```python
country=["india","germenay","usa","france","africa"]
gdp=[25355,46636,34646,77377,43333]
plt.bar(country,gdp)
plt.title("country gdp using barplot")
plt.xlabel("country")
plt.ylabel("gdp")
plt.show()
```

```python
plt.barh(country,gdp)
plt.title("country gdp using barplot")
plt.xlabel("country")
plt.ylabel("gdp")
plt.show()
```

```python
num=[0,2,3,4,5,6,7,8,5]
plt.hist(num,bins=3)
plt.xlabel("number")
plt.ylabel("histogram frequency")
plt.title("histogram")
plt.show()
```

```python
plt.hist(num,bins=3,histtype="bar")
plt.xlabel("number")
plt.ylabel("histogram frequency")
plt.title("histogram")
plt.show()
```

```python
y=[20,34,54,25]
plt.pie(y)
plt.show()
```

```python
y=[20,34,25]
color=["r","b","green"]
plt.pie(y,colors=color,labels=["dsp","wt","coa"])
plt.legend()
plt.show()
```

```python
x=[0,1,2,3]
y=[3,8,6,8]
plt.subplot(1,2,1)
plt.plot(x,y)
y_=[23,43,45,22]
plt.subplot(1,2,2)
plt.plot(x,y_)
plt.show()
```

```python
x=[0,1,2,3]
y=[3,8,6,10]
plt.subplot(2,2,1)
plt.plot(x,y)
plt.subplot(2,2,2)
plt.bar(x,y,color="y")
y=[23,30,10,45]
plt.subplot(2,2,3)
plt.pie(y)
y_=[3,8,6,10]
plt.subplot(2,2,4)
plt.bar(x,y_,color="red")
plt.show()
```
