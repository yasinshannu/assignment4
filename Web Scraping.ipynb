```python
import urllib.request

url="https://www.rguktsklm.ac.in/"
resp=urllib.request.urlopen(url)
page=str(resp.read())


def findurl(page):
    start=page.find("<a href")
    start=page.find('"',start)
    end=page.find('"',start+1)
    url=page[start+1:end]
    return url,end

while True:
    url,n=findurl(page)
    page=page[n:]
    if url:
        print(url)
    else:
        break
    
```
